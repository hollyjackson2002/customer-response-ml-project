## Notebook 1: Exploratory Data Analysis
This notebook covers dataset description, missing values, descriptive statistics, and graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('../data/raw/marketing_campaign.csv')
FIG_DIR = Path('../figures')
FIG_DIR.mkdir(exist_ok=True)

df = pd.read_csv(DATA_PATH, sep='\t')
df.head()
```

### Dataset shape and structure
Use this output to describe how many rows and columns the dataset has.

```python
print('Rows and columns:', df.shape)
df.info()
```

### Descriptive statistics
Use this table in the EDA section to discuss average income, spending and customer behaviour.

```python
df.describe().T
```

### Missing data analysis
This checks which variables have missing values.

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values[missing_values > 0]
```

### Basic data cleaning for EDA
Income has missing values, so these rows are removed for reliable analysis.

```python
df = df.dropna().copy()
df['Age'] = 2025 - df['Year_Birth']
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
print('Cleaned dataset shape:', df.shape)
```

### [Campaign Response Distribution]
This graph shows whether the target variable is balanced or imbalanced.

```python
plt.figure(figsize=(6,4))
df['Response'].value_counts().sort_index().plot(kind='bar')
plt.title('Campaign Response Distribution')
plt.xlabel('Response: 0 = No, 1 = Yes')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.savefig(FIG_DIR / 'campaign_response_distribution.png', dpi=300)
plt.show()

df['Response'].value_counts(normalize=True) * 100
```

### [Income Distribution]
This graph helps explain customer income variation and outliers.

```python
plt.figure(figsize=(6,4))
df['Income'].hist(bins=30)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig(FIG_DIR / 'income_distribution.png', dpi=300)
plt.show()
```

### [Customer Age Distribution]
This graph shows the age spread of customers in the dataset.

```python
plt.figure(figsize=(6,4))
df['Age'].hist(bins=25)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig(FIG_DIR / 'customer_age_distribution.png', dpi=300)
plt.show()
```

### [Total Spending vs Campaign Response]
This graph compares spending behaviour between customers who accepted and rejected the campaign.

```python
plt.figure(figsize=(6,4))
df.boxplot(column='Total_Spending', by='Response')
plt.title('Total Spending vs Campaign Response')
plt.suptitle('')
plt.xlabel('Response')
plt.ylabel('Total Spending')
plt.tight_layout()
plt.savefig(FIG_DIR / 'total_spending_vs_response.png', dpi=300)
plt.show()
```

### [Correlation Heatmap]
This graph shows relationships between numerical features.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12,8))
plt.imshow(numeric_df.corr(), aspect='auto')
plt.colorbar()
plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=90)
plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig(FIG_DIR / 'correlation_heatmap.png', dpi=300)
plt.show()
```
